```python
from IPython.display import display,Math
import sympy as sym
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x = 3
y = 5

plt.plot(x,y,'ro') # r indicates red, o indicates circle
# other options. 'gs' makes green square. 'kp' makes black pentagon

plt.axis('square') # makes square
plt.axis([-6,6,-6,6]) # limits
plt.grid() # shows grid

plt.show() # shows final version of figure
```

```python
# demonstrates that order matters
x = 3
y = 5

plt.plot(x,y,'ro') # r indicates red, o indicates circle
# other options. 'gs' makes green square. 'kp' makes black pentagon

plt.axis([-6,6,-6,6]) # limits
plt.axis('square') # makes square
plt.grid() # shows grid

plt.show() # shows final version of figure
```

```python
x = [-4,5,2,4,5,6,3]
y = [0,3,4,5,2,-3,4]

for i in range(0,len(x)):
    plt.plot(x[i],y[i],'o',label='point %s' %(1+i))

plt.axis('square')
plt.grid()
plt.legend() # used for labeling points (ie.label='qwerty' or label='point %s' %(1+i) )
plt.show()
```
