```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import display,Math
%matplotlib inline
from PIL import Image
from matplotlib.patches import Polygon
```

```python
# y = 2 - x**2
x = range(-3,4)
y = np.zeros(len(x))

for i in range(0,len(x)):
    y[i] = 2 - x[i]**2

plt.plot(x,y, 's-')
plt.show()
```

```python
# resolution (number of points)

x = np.linspace(-3,3,10) # first input is upper bound, second input is lower bound, third input is number of steps
y = 2 + np.sqrt(abs(x))

plt.plot(x,y,'ms-')
plt.show()
```

```python
# resolution (number of points) (continued)

x = np.linspace(-3,4,4) # first input is upper bound, second input is lower bound, third input is number of steps
y1 = 2 + abs(x)**(1/3)
y2 = 2 + abs(x)**(1/2)

plt.plot(x,y1,'m-')
plt.plot(x,y2, 'k')
plt.show()
```

```python
# exercise.indepentent
x,y = sym.symbols('x,y')
x = np.linspace(-1,3,20)
# xi = -1
for i in range(-1,4):
    y = x**i
    plt.plot(x,y, label='y=x%s'%(i))

plt.axis([-4,4,-20,20])
plt.legend()
plt.show()
# display(Math(sym.sympify(x)))
```

```python
# exercise.instructed
x = np.linspace(-4,4,100)
e = range(-1,4)

for i in e:
    y = x**i
    plt.plot(x,y,label='$y=x^{%s}$' %i,linewidth=4) # encased in $ converts to latex, x^{%s} show latex exponent
    
plt.legend()
plt.xlim([x[0],x[-1]])
plt.ylim([-20,20])

# plt.savefig('coolPythonGraph.pdf') # exports as figure

plt.show()
```
