```python
# graphing lines part 2
# slope/intercept form (ie: y = mx + b)
# m = "slope" (rise over run)
# b = "intercept" (where line goes over x axis when the y axis = 0(when y = 0, x = b))
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# y = mx + b ( draw line using formula )

x = [-5,5]

m = 2
b =1

y = [340,2345] # 2 element vector for y
for i in range(0,len(x)):
    y[i] = m*x[i]+b
    
plt.plot(x,y,label='y=%sx+%s'%(m,b))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')

plt.legend()

plt.show()
```

```python
# y = mx + b / draw line using formula
# converts x into numpy array

x = [-5,5]

m = 2
b =1

# y = [340,2345] # 2 element vector for y
# for i in range(0,len(x)):
#     y[i] = m*x[i]+b

m = m*np.array(x) + b
    
plt.plot(x,y,label='y=%sx+%s'%(m,b))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')

plt.legend()

plt.show()
```

```python
# exercise.independent
# y = mx + b / draw line using formula

x = [-5,5]

mb1 = [.7,-2]
mb2 = [-1.25,.75]

y = [0,0] # 2 element vector for y
for i in range(0,len(x)):
    y[i] = mb1[0]*x[i]+mb1[1]

plt.plot(x,y,label='y=%sx+%s'%(mb1[0],mb1[1]))

for i in range(0,len(x)):
    y[i] = mb2[0]*x[i]+mb2[1]
plt.plot(x,y,label='y=%sx+%s' %(mb2[0],mb2[1]))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')

plt.legend()

plt.show()
print(len(x))
```

```python
# exercise.instructed
x = [-5,5]
m = [.7,-1.25]
b = [-2,.75]

for i in range(0,len(m)):
    y = m[i]*np.array(x) + b[i]
    plt.plot(x,y,label='%sx + %s' %(m[i],b[i]))
    
plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')

plt.legend()
plt.show()
```
