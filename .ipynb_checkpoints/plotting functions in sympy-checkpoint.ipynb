```python
import sympy as sym
import sympy.plotting.plot as symplot
import numpy as np
```

```python
from sympy.abc import x

y = x**2

p = symplot(y,show=False)
p.xlim = [0,50]
p[0].line_color = 'm'
p.title = 'this is a nice plot'

p.show()
```

```python
x,a = sym.symbols('x,a')

expr = a/x

n = 1
n2 = 2

p = symplot(expr.subs(a,n),(x,-5,5),show=False)
p[0].label = '$y=%s$' %sym.latex(expr.subs(a,n)) # encase in $ (ie '$y=%s' ) ilistrates inside string is latex

p.extend( symplot(expr.subs(a,n2),show=False) ) # extend by 1 more element 
p[1].label = '$y=%s$' %sym.latex(expr.subs(a,n2)) # encase in $ (ie '$y=%s' ) ilistrates inside string is latex


p[0].line_color = 'r'
p.ylim = [-5,5]
p.legend = True
p.show()
```
