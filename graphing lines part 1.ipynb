```python
from IPython.display import display,Math
import numpy as np
import sympy as sym
import math
import matplotlib.pyplot as plt
```

```python

p1 = [-3,-1]
p2 = [4,4]

plt.plot([p1[0],p2[0]],[p1[1],p2[1]],color=[1,0,0],linewidth=1) # color=[1,0,0] means RGB 100% red, 0% green, 0% blue

plt.axis('square')
plt.axis([-6,6,-6,6])
plt.grid()
plt.show()
```

```python

x = 3
y = 5
plt.plot(x,y,'ro')
plt.plot([0,x],[0,y]) # plots straight line  

plt.axis('square')
plt.axis([-6,6,-6,6])
plt.grid()

plt.plot([-6,6],[0,0], 'k') # horizontal line at x=0
plt.plot([0,0],[-6,6], 'k') # vertical line at y=0

plt.show()
```

```python
# first exercise.independent

x = sym.symbols('x')
y = sym.sqrt(sym.Abs(x))

for pnti in range(-20,21):
    plt.plot([0,pnti],[0,y.subs(x,pnti)])

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# first exercise.instructed

x = range(-20,21)

for xi in x:
    plt.plot([0,xi],[0,abs(xi)**(1/2)]) # sqrt is n**(1/2) # can also use np mod np.sqrt(xi)
    
plt.show()
```

```python
# experimental
# first exercise.independent

x = sym.symbols('x')
y = sym.sqrt(sym.Abs(x))

for pnti in range(-20,21):
    plt.plot([0,pnti],[0,y.subs(x,pnti)])

plt.xlabel('x')
plt.ylabel(sym.latex(sym.sympify(y)))
plt.show()
```
