```python
import numpy as np
import sympy as sym
import math
from IPython.display import display,Math
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
```

```python
y = np.array([ [1,1], [2,3], [3,1] ] )
p = Polygon(y,facecolor='m',alpha=.3)

y1 = np.array([ [2,2], [2.5,4], [3.5,1] ] )
p1 = Polygon(y1,facecolor='b',alpha=.3,edgecolor='k')

fig,ax = plt.subplots()

ax.add_patch(p)
ax.add_patch(p1)

ax.set_ylim(0,4)
ax.set_xlim(0,4)
plt.show()
```

```python
# exercise.independent
x = np.linspace(-2, 2, 1000)

y = -x**2

fig, ax = plt.subplots()
# ax.plot(y, x)
ax.plot(x, y)

y = np.array([ [-.5,-4], [-.5,-2.5], [.5,-2.5], [.5,-4] ] )
p = Polygon(y,facecolor='k',alpha=1)

fig,ax = plt.subplots()

ax.add_patch(p)

ax.set_ylim(-4,0)
ax.set_xlim(-2,2)
plt.show()
```

```python
# exercise.instructed
x = np.linspace(-2,2,101)
f = -x**2

pnts = np.vstack((x,f)).T
p = Polygon(pnts, facecolor='y',alpha=.2, edgecolor='k')

y = np.array([ [-.5,-4], [-.5,-2.5], [.5,-2.5], [.5,-4] ] )
p1 = Polygon(y, facecolor='k',alpha=1, edgecolor='k')

fig,ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)

plt.plot(x,f,'k')
plt.plot([-2,2],[-4,-4],'k')
plt.show()
```
